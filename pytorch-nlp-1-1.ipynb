```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
corpus=['Time flies like an arrow.','Fruit flies like a banana.'] # sentence 1,sentence 2
one_hot_vectorizer=CountVectorizer(binary=True) # set this class and set the instance
one_hot=one_hot_vectorizer.fit_transform(corpus).toarray() # put the corpus in and change to array
vocab=one_hot_vectorizer.get_feature_names() # get all the words or so called vocab
sns.heatmap(one_hot,annot=True,cbar=False,xticklabels=vocab,yticklabels=['Sentence 1','Sentence 2'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns

tfidf_vectorizer=TfidfVectorizer()
tfidf=tfidf_vectorizer.fit_transform(corpus).toarray()
vocab=tfidf_vectorizer.get_feature_names()
sns.heatmap(tfidf,annot=True,cbar=True,xticklabels=vocab,yticklabels=['Sentence 1','Sentence 2'])
```

```python
def describe(x):
    print("Type:{}".format(x.type()))
    print("Shape/size{}".format(x.shape))
    print("Values: \n{}".format(x))
```

```python
import torch
describe(torch.Tensor(2,3)) # input Torch.tensor(2,3)
```

```python
import torch 
describe(torch.rand(2,3))
describe(torch.randn(2,3)) # normalization,mean=0,variance=1
```

```python
import torch
describe(torch.zeros(2,3)) # torch.zeros
x=torch.ones(2,3) # torch.ones
describe(x)
x.fill_(5)
describe(x)
```

```python
# use python list to declare a tensor
x=torch.Tensor([[1,2,3],[4,5,6]]) # defalue are float tensor
describe(x)
```
