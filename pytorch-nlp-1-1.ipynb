```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
corpus=['Time flies like an arrow.','Fruit flies like a banana.'] # sentence 1,sentence 2
one_hot_vectorizer=CountVectorizer(binary=True) # set this class and set the instance
one_hot=one_hot_vectorizer.fit_transform(corpus).toarray() # put the corpus in and change to array
vocab=one_hot_vectorizer.get_feature_names() # get all the words or so called vocab
sns.heatmap(one_hot,annot=True,cbar=False,xticklabels=vocab,yticklabels=['Sentence 1','Sentence 2'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import seaborn as sns

tfidf_vectorizer=TfidfVectorizer()
tfidf=tfidf_vectorizer.fit_transform(corpus).toarray()
vocab=tfidf_vectorizer.get_feature_names()
sns.heatmap(tfidf,annot=True,cbar=True,xticklabels=vocab,yticklabels=['Sentence 1','Sentence 2'])
```

### 基於維度的 tensor 運算

```python
import torch
x=torch.arange(6)
describe(x)
```

```python
x=x.view(2,3) # view(),permute(),both are change the dimension of the tensor
describe(x)
```

```python
describe(torch.sum(x,dim=0)) # col相加
```

```python
describe(torch.sum(x,dim=1)) # row相加
```

```python
describe(torch.transpose(x,0,1)) # torch.transpose 轉置矩陣
```

### 索引,分割,連結

```python
import torch
x=torch.arange(6).view(2,3)
describe(x)
```

```python
describe(x[:1,:2]) 
```

```python
describe(x[0,1]) # x[row,col]
```

```python
# non-continous index
indices=torch.LongTensor([0,2])
print(indices)
describe(torch.index_select(x,dim=1,index=indices))
```

```python
indices=torch.LongTensor([0,0])
describe(torch.index_select(x,dim=0,index=indices))
```

```python
# 連結tensor
import torch
x=torch.arange(6).view(2,3)
describe(x)
```

```python
describe(torch.cat([x,x],dim=0)) # dim=0 -> col
```

```python
describe(torch.cat([x,x],dim=1)) #dim=1 -> row
```

```python
# tensor 的線性代數
import torch 
x1=torch.arange(6).view(2,3)
describe(x1)
```

```python
x2=torch.ones(3,2)
x2[:,1]+=1 #  第二個col加上1
describe(x2)
```

```python
x2=x2.long()
describe(torch.mm(x1,x2)) # 矩陣相乘
```

## Tensor 與 計算圖

```python
import torch
x=torch.ones(2,2,requires_grad=True)
describe(x)
print(x.grad is None)
```

```python
y=(x+2)*(x+5)+3
describe(y)
print(x.grad is None)
```

```python
z=y.mean()
describe(z)
z.backward()
print(x.grad is None)
```

## 建構CUDA Tensor

```python
import torch
print(torch.cuda.is_available())
```

```python
# preferred method:device agnostic tensor instantiation
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
x=torch.rand(3,3,).to(device) #利用to(device),將tensor的計算移動到gpu上
describe(x)
```

```python
#混用cuda & cpu的tensor
y=torch.rand(3,3)
x+y
```

```python
cpu_device=torch.device("cpu")
y=y.to(cpu_device)
x=x.to(cpu_device)
x+y
```
