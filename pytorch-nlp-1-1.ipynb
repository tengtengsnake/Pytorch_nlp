```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns
corpus=['Time flies like an arrow.','Fruit flies like a banana.'] # sentence 1,sentence 2
one_hot_vectorizer=CountVectorizer(binary=True) # set this class and set the instance
one_hot=one_hot_vectorizer.fit_transform(corpus).toarray() # put the corpus in and change to array
vocab=one_hot_vectorizer.get_feature_names() # get all the words or so called vocab
sns.heatmap(one_hot,annot=True,cbar=False,xticklabels=vocab,yticklabels=['Sentence 1','Sentence 2'])
```
